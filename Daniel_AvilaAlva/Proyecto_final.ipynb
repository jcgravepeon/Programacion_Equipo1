```
import pandas as pd
import matplotlib.pyplot as plt
import glob
import matplotlib.dates as mdates
import numpy as np

```
Importo todas las librerias que voy a utilizar 


```
listafile=glob.glob("2023*.csv")
```
se crea una variable para enlistar todos los archivos que tengan la terminación **"2023.csv"**

```
data=[]
for files in listafile:
  data.append(pd.read_csv(files))
datafull=pd.concat(data,ignore_index=True)
```

Se crea otra variable con un ***for*** para porder juntar la variable **data** con la variable **listafiles** para que sea una variable que lea todos los achivos de la lista.

Para que despues se hiciera una variable concatentando la variable **data** para ponerla en formato de tabla y que ignore los index, y asi es que ya tenmos una variable con todos los archivos juntos organizados en una tabla.

```
Time=pd.to_datetime(datafull["Datetime"][:],format="%Y-%m-%d %H:%M:%S")
```
Se crea una variable para ponerle un formato de ***año,mes,dia,hora,minuto y segundo*** a los datos de la columna ***Datetime***.

```
datafull2=datafull.drop("Datetime",axis=1)
```
Se crea una variable donde tengamos la tabla de **datafull** pero donde elimine la columna llamada ***Datetime***   

```
datafull2.insert(0,"Time",Time,True)
```
Se agrega el **Time** a la columna que tenia el ***Datetime*** 

```
datagood=datafull2.sort_values(by="Time",ignore_index=True)
```
Aqui se ordenan los datos conforme a la marca de Tiempo

```
datagood['Presion'].replace(0,np.nan,inplace=True)
```
Aqui se remplaza los 0 por nan en la columna ***"Presion"***

```
fig,(ax1)=plt.subplots(figsize=(15,5))
ax1.set_xlabel("Tiempo")
ax1.set_ylabel("Temperatura")
ax1.plot(datagood['Time'],datagood[' Temperatura'],label="Temperatura",color="r")
ax1.plot(datagood['Time'],datagood['Humedad'],label="Humedad",color="g")
#fig.autofmt_xdate()
ax2=ax1.twinx()  #Aqui pusimos que la x (Tiempo) sea la misma para las 2 graficas
ax2.set_ylabel("Presion")
ax2.plot(datagood['Time'],datagood['Presion'],label=("Humedad"),color="b")
```
