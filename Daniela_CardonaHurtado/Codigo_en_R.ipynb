# INTERACCIONES

Utilizacion de liberias matplotlib y networkx.
Utilizacion del modulo bipartite de networkx para la proyeccion y analizar las similitudes en dos conjuntos.

```python
import networkx as nx
import matplotlib.pyplot as plt
from networkx.algorithms import bipartite
```

```python
IN = nx.Graph()
```

```python
fig, ax1 = plt.subplots(figsize=(16,10))
IN = nx.Graph()
IN.add_nodes_from(['HABITAT','ALIMENTO'], bipartite=0)
IN.add_nodes_from(['Escarabajo Minero','Escarabajo Gorgojo','Araña Beige','Arañas Patas Largas','Gluvia Dorsalis','Araña Verde','Hormiga Negra','Mariposa','Orthoptero Verde','Grillo','Araña camello','Araña negra con franjas Amarillas','Araña Negra','Abeja Melera','Insecto Palo'],bipartite=1)
IN.add_edges_from([('Alimento','Escarabajo minero'),('Alimento','Escarabajo Gorgojo'),('Alimento','Arañas patas largas'),('Alimento','Gluvia Dorsalis'),('Alimento','Araña verde'),('Alimento','hormiga negra'),('Alimento','Mariposa'),('Alimento','Orthoptero Verde'),('Alimento','Grillo'),('Alimento','Araña Camello'),('Alimento','Araña negra con franjas amarillas'),('Alimento','Araña negra'),('Alimento','Abeja Melera'),('Alimento','Insecto Palo')])
IN.add_edges_from([('Habitat','Escarabajo Minero'),('Habitat','Escarabajo Gorgojo'),('Habitat','Araña Beige'),('Habitat','Arañas Patas Largas'),('Habitat','Araña Verde'),('Habitat','Araña Camello'),('Habitat','Insecto Palo')])
nx.draw_networkx(IN, pos = nx.drawing.layout.bipartite_layout(IN, ['Habitat','Alimento']), width = 2, node_color= "orange")
```
