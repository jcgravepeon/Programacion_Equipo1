# Red de interacciones 

## Emular la red de interacciones hecha en R el semestre pasado ahora en python.

```python
import networkx as nx
import matplotlib.pyplot as plt
from networkx.algorithms import bipartite
```

```python
A=nx.Graph()
```

```python
A.add_nodes_from(['Habitat','Alimento'], bipartite=0)
A.add_nodes_from(['Escarabajo minero','Escarabajo Gorgojo','Araña beige','Arañas patas largas','Gluvia Dorsalis','Araña verde','hormiga negra','Mariposa','Orthoptero verde','Grillo','Araña camello','Araña negra con franjas amarillas','Araña negra','Abeja Melera','Insecto Palo'],bipartite=1)
```

```python
A.add_edges_from([('Alimento','Escarabajo minero'),('Alimento','Escarabajo Gorgojo'),('Alimento','Arañas patas largas'),('Alimento','Gluvia Dorsalis'),('Alimento','Araña verde'),('Alimento','hormiga negra'),('Alimento','Mariposa'),('Alimento','Orthoptero verde'),('Alimento','Grillo'),('Alimento','Araña camello'),('Alimento','Araña negra con franjas amarillas'),('Alimento','Araña negra'),('Alimento','Abeja Melera'),('Alimento','Insecto Palo')])
A.add_edges_from([('Habitat','Escarabajo minero'),('Habitat','Escarabajo Gorgojo'),('Habitat','Araña beige'),('Habitat','Arañas patas largas'),('Habitat','Araña verde'),('Habitat','Araña camello'),('Habitat','Insecto Palo')])
```

```python
nx.draw_networkx(A, pos = nx.drawing.layout.bipartite_layout(A, ['Habitat','Alimento']), width = 2, font_size=10, style='dashed', node_shape='s')
```
