# Código de atmósfera 

Éste esquema muestra la atmósfera y su temperatura respecto a la distancia recorrida en el eje y, lo que muestra es la disminución de la temperatura a medida que incrementa la altitud , al subir nuevamente ene una capa intermedia indica  que hay otra fuente de calor en una capa más arriba. 

```python
import numpy as np  # librería para conjuntos  operaciones matemáticas
import matplotlib.pyplot as plt #librería para poder realizar gráficas 
from ambiance import Atmosphere #Ambiance es una implementación de datos estándar de la Atmósfera

```

```python
# Creando variables 
heights = np.linspace(-5e3, 80e3, num=1000) #se establece el rango de -5 a la menos tres hasta 80 a la 3 en la altura de acuerdo a los km
atmosphere = Atmosphere(heights)#variable de la atmósfera respecto a la altura 
```

```python
# se plotean las variables respecto a "x" y "y" 
plt.plot(atmosphere.temperature_in_celsius, heights/1000) #la temperatura en grados celsius y se divide la altura respecto a los kilometros
plt.ylabel('Height [km]') #graficación respecto a la altura en el eje y 
plt.xlabel('Temperature [°C]') #graficación respecto a la altura en el eje x
plt.grid() # muestra las líneas del gráfico
plt.show() # especifica que el gráfico sea mostrado
```

Se puede observar en la figura la disminución de la temperatura y su incremento lo que indica la capa donde el 03 se encuentra, es la capa atmosférica donde se filtra mayormente los rayor uv, debido a reacciones en donde las moléculas se disocian, debido a estos procesos en esa capa de la atmósfera la temperatura incrementa.
